```python
def get_chrome_driver(headless_flag):
    # Set up the Selenium webdriver
    from selenium import webdriver
    from selenium.webdriver.chrome.options import Options
    from selenium.webdriver.chrome.service import Service

    service = Service(executable_path=DRIVER_FILE_PATH)
    chrome_options = Options()
    
    if headless_flag:
        # in case you want headless browser
        
        chrome_options.add_argument("--headless")
        chrome_options.add_argument("--disable-extensions")
        chrome_options.add_argument("--disable-gpu")
        chrome_options.add_experimental_option('prefs', {'headers': headers})
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--start-maximized")
        chrome_options.add_argument("user-agent={}".format(random.choice(USER_AGENT_LIST)))
    
    driver = webdriver.Chrome(service=service, options=chrome_options) 

    return driver
```

```python
from selenium.webdriver.common.by import By
def get_card_data(card): 
    try:
        activity_type = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[1]/header/div/span').text
    except Exception as e:
        activity_type = ''
    try:
        activity_name = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[1]/header/p').text
    except Exception as e:
        activity_name = ''
    try:
        activity_time = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[1]/div[2]/ul/li[1]/span').text
    except Exception as e:
        activity_time = ''
    try:
        activity_addtitional = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[1]/div[2]/ul/li[2]/span').text
    except Exception as e:
        activity_addtitional = ''
    try:
        activity_price = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[2]/footer/div[2]/div/div[1]').text
    except Exception as e:
        activity_price = ''
    try:
        activity_discount_price = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div[11]/article/div/a/div[2]/footer/div[2]/div/div[1]/p').text
    except Exception as e:
        activity_discount_price = ''
    try:
        activity_rating = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[2]/footer/div[1]/div[1]/span').text
    except Exception as e:
        activity_rating = ''
    try:
        activity_total_review = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[2]/footer/div[1]/div[2]/span').text
    except Exception as e:
        activity_total_review = ''
    try:
        activity_image_link = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a/div[1]/div[1]/div/picture/img').get_attribute("src")
    except Exception as e:
        activity_image_link = ''
    try:
        activity_link = card.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div/div/div/div[1]/section[1]/div/article/div/a').get_attribute("href")
    except Exception as e:
        activity_link = ''

    result = {
        "activity_type": activity_type,
        "activity_name": activity_name,
        "activity_time": activity_time,
        "activity_addtitional": activity_addtitional,
        "activity_price": activity_price,
        "activity_discount_price": activity_discount_price,
        "activity_rating": activity_rating,
        "activity_total_review": activity_total_review,
        "activity_image_link": activity_image_link,
        "activity_link": activity_link
    }
    return result
```

```python
# Function to check if the page content has changed after clicking the "Next" button
def has_page_content_changed(previous_content):
    current_content = driver.page_source
    return current_content != previous_content

while True:
    grid_wrapper = driver.find_element(By.CSS_SELECTOR, ".grid-wrapper")
    all_cards = grid_wrapper.find_elements(By.CSS_SELECTOR, ".gtm-trigger__card-interaction")
    previous_data_count = len(all_cards)
    previous_page_content = driver.page_source

    for card in all_cards:
        data_item = get_card_data(card)
        # ingest this data into mongodb
        collection.insert_one(data_item)

    try:
        time.sleep(1)
        next_button = driver.find_element(By.CSS_SELECTOR, ".next")
        driver.execute_script("arguments[0].click();", next_button)
        print("Next button is clicked.")
        
        if not has_page_content_changed(previous_page_content):
            print("No new data loaded. Exiting the loop.")
            break
    except Exception as e:
        print("Next button is not available.", e)
        break  # break the while loop
```
