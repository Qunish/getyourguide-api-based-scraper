```python
for page in range(1,51):
    try:
        json_data['events'][0]['event']['payload']['page'] = page
        response = requests.post(
                    'https://travelers-api.getyourguide.com/user-interface/landing-pages/blocks',
                headers=headers,
                json=json_data
            )
        response_json = json.loads(response.text)
        items = response_json['content'][0]['content']
        
        print("Scraping data for page:", page)
        
        result = collection.insert_many(items)
        time.sleep(random.choice([3,2]))
        # print(json_data)
    except Exception as e:
        print("Exception occured: ",e)
```

```python
# get all the lv data
data = collection.find({})
```

```python
data_result = list(data)
```

```python
import pandas as pd
df = pd.DataFrame(data_result)
```

```python
len(df)
```

```python
df['activity_url'] = df['onClickLink'].apply(lambda x: "https://www.getyourguide.com" + x['link'])
df['activity_price'] = df['price'].apply(lambda x: x['prefix'] + " " + x['startPrice'] + " " + x['suffix'])
for index, activity in df.iterrows():
    df.loc[index, 'activityTypeLabel'] = activity['activityTypeLabel']['text']
    df.loc[index, 'rating'] = activity['rating']['formattedRating']
df.drop_duplicates('activity_url',inplace=True)
len(df)
```
